**Question 2: What are the types of machine learning? Describe each with one
real-world example.**

Answer - Types of machine learning are:
*   Supervised
*   UnSupervised
*   Semi-Supervised
*   Reinforcement

**1.Supervised**

*  Historical data act as supervisor
*  Where dependent variables (Y) is present.


**Example:**

Email Spam Detection – The algorithm is trained with a dataset of emails labeled as “spam” or “not spam.” Once trained, it can classify new incoming emails accordingly.



**2.Unsuprvised**

*  No supervision
*  No any historical data

**Example:**

Customer Segmentation – E-commerce companies group customers based on buying behavior using clustering techniques.

**3.Semi-Supervised Learning**

It's mix of supervised and unsupervised.

**Example:**

Medical Diagnosis with Limited Data – Hospitals may have a small number of patient scans labeled with a disease type but a large amount of unlabeled scans. Semi-supervised learning can improve diagnosis accuracy.

**4. Reinforcement**

The model (agent) learns by interacting with an environment. It takes actions, receives feedback in the form of rewards or penalties, and improves its strategy over time.

**Example**


Self-Driving Cars – The AI learns to navigate by receiving rewards for staying in the lane and avoiding collisions and penalties for crashes or traffic violations.




**Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.**

Answer -

1. Overfitting

Overfitting happens when a machine learning model learns too much from the training data, including noise and irrelevant details.

As a result, it performs extremely well on training data but poorly on new, unseen data.

2. Underfitting

Underfitting occurs when a model is too simple to capture the underlying patterns in the data. It performs poorly on both training and test data.

3. Bias–Variance Tradeoff

The bias–variance tradeoff explains the balance between bias (error due to overly simplistic assumptions in the model) and variance (error due to model sensitivity to small changes in the training data).

High bias → underfitting.

High variance → overfitting.

**Question 4: What are outliers in a dataset, and list three common techniques for
handling them.**

Answer - Outliers are the data points that significantly differ from the majority of observations in a dataset.

They may be unusually high or low compared to the rest of the data.

**Three Common Techniques for Handling Outliers**

1.  Removal of Outliers

If the outlier is due to measurement error or is clearly irrelevant, it can be removed.

2. Transformation of Data

Apply transformations like logarithmic, square root, or Box–Cox to reduce the effect of extreme values.

3.  Capping (Winsorizing)

Limit extreme values to a certain percentile range (e.g., 1st to 99th percentile).

**Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data**

Answer - Process of Handling Missing Values

1. Identify Missing Values

Use methods like isnull() or isna() in Python (Pandas) to locate missing entries.

2.  Determine the Cause

Missing values can occur due to data entry errors, sensor failures, skipped survey questions, etc.

3.  Decide on a Strategy

Remove missing values if they are few and random.

 Impute (fill in) missing values if they are significant and contain useful patterns.

4. Apply the Chosen Method

Use appropriate imputation depending on whether the data is numerical or categorical.

5. Validate Results

After imputation, check if the changes maintain data integrity and do not introduce bias.

**Imputation Techniques**

For Numerical Data → Mean Imputation

Replace missing values with the mean of the available values.

Example: If the ages in a dataset are [25, 28, NaN, 32], replace NaN with (25 + 28 + 32)/3 = 28.33.

For Categorical Data → Mode Imputation

Replace missing values with the most frequent category.

Example: If the colors are [Red, Blue, NaN, Red], replace NaN with "Red" (most common).

Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(len(data), size=50, replace=False)
data[missing_indices] = np.nan

# Create DataFrame
df = pd.DataFrame({'Values': data})

# Step 3: Save histogram before imputation
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(df['Values'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Step 4: Fill missing values with column mean
mean_value = df['Values'].mean()
df['Values'] = df['Values'].fillna(mean_value)

# Step 5: Save histogram after imputation
plt.subplot(1, 2, 2)
plt.hist(df['Values'], bins=30, color='lightgreen', edgecolor='black')
plt.title("After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Question 10: You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:

● Missing ages,

● Outliers in transaction amount,

● A highly imbalanced target (fraud vs. non-fraud),

● Categorical variables like payment method.


```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE

# Create a synthetic dataset with relevant columns
np.random.seed(42) # for reproducibility

n_samples = 1000
# Create a slightly imbalanced classification dataset
X, y = make_classification(n_samples=n_samples, n_features=5, n_informative=3, n_redundant=0,
                           n_clusters_per_class=1, weights=[0.9, 0.1], flip_y=0.01, random_state=42)

# Convert to DataFrame
df_synthetic = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(X.shape[1])])
df_synthetic['Age'] = np.random.randint(18, 70, size=n_samples)
df_synthetic['TransactionAmount'] = np.random.uniform(10, 500, size=n_samples)
df_synthetic['PaymentMethod'] = np.random.choice(['CreditCard', 'DebitCard', 'PayPal', 'Other'], size=n_samples)
df_synthetic['IsFraud'] = y # The target variable

# Introduce missing values in 'Age'
missing_age_indices = np.random.choice(n_samples, size=50, replace=False)
df_synthetic.loc[missing_age_indices, 'Age'] = np.nan

# Introduce outliers in 'TransactionAmount'
outlier_indices = np.random.choice(n_samples, size=10, replace=False)
df_synthetic.loc[outlier_indices, 'TransactionAmount'] = np.random.uniform(1000, 5000, size=10)

print("Synthetic DataFrame created:")
display(df_synthetic.head())
print("\nClass distribution of the target variable:")
print(df_synthetic['IsFraud'].value_counts())
```

```python
# 1. Handle Missing Ages (using median imputation)
df_processed = df_synthetic.copy() # Create a copy to avoid modifying the original synthetic data
df_processed['Age'] = df_processed['Age'].fillna(df_processed['Age'].median())
print("Missing values in 'Age' after imputation:")
print(df_processed['Age'].isnull().sum())

# 2. Handle Outliers in Transaction Amount (using IQR method)
Q1 = df_processed['TransactionAmount'].quantile(0.25)
Q3 = df_processed['TransactionAmount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_processed_no_outliers = df_processed[(df_processed['TransactionAmount'] >= lower_bound) &
                                        (df_processed['TransactionAmount'] <= upper_bound)].copy()
print("\nShape of DataFrame after outlier removal:")
print(df_processed_no_outliers.shape)


# 3. Handle Imbalanced Target (using SMOTE)
# Separate features and target for SMOTE
X_smote = df_processed_no_outliers.drop('IsFraud', axis=1)
y_smote = df_processed_no_outliers['IsFraud']

# Need to handle categorical features before SMOTE if any are present in X_smote
# In this synthetic example, 'PaymentMethod' is still in X_smote and is categorical.
# We will one-hot encode it before applying SMOTE.

X_smote = pd.get_dummies(X_smote, columns=['PaymentMethod'], drop_first=True)


smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_smote, y_smote)

print("\nClass distribution after SMOTE:")
print(pd.Series(y_res).value_counts())

# 4. Handle Categorical Variables (One-Hot Encoding - already done for SMOTE)
# If you were not using SMOTE immediately after outlier removal, you would
# perform one-hot encoding on the relevant categorical columns here.
# Example:
# df_processed_encoded = pd.get_dummies(df_processed_no_outliers, columns=['PaymentMethod'], drop_first=True)
# print("\nDataFrame after one-hot encoding (example if not using SMOTE):")
# display(df_processed_encoded.head())

print("\nData handling steps demonstrated on synthetic data.")
```
